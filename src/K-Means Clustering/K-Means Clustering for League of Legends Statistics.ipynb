```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
import numpy as np
from sklearn.model_selection import train_test_split

DATA_DIR = '../../data/'
print(os.listdir(DATA_DIR))
```

```python
filename = DATA_DIR + 'CleanLoLMatchHistory2.csv'
df = pd.read_csv(filename)
df
```

```python
def parse_gold(x):
    if('k' in x):
        x = x.replace('k','')
        return float(x) * 1000
    else:
        return float(x)

df.columns = df.columns.str.lower().str.replace(' ', '_')
df['gold'] = df['gold'].apply(parse_gold)
```

```python
# df.info()
#df_parsed = df[['level','creep_score','gold','experience_gained','kills','deaths','assists','game_status']]
#df_parsed = df[['kills','deaths','assists','level','game_status',]]
df_parsed = df[['kills','deaths','assists','level','game_status',]]
df_parsed.head()
```

```python
df_parsed.head()
```

```python
def get_status(n):
    if n == 1:
        return 'Win'
    elif n == 0:
        return 'Draw'
    elif n == -1:
        return 'Lose'
    
df_parsed['game_status'] = df_parsed['game_status'].apply(get_status)
df_parsed.head()
```

```python
# KNN
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 3)
x,y = df_parsed.loc[:,df_parsed.columns != 'game_status'], df_parsed.loc[:,'game_status']
knn.fit(x,y)
prediction = knn.predict(x)
print('Prediction: {}'.format(prediction))
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3,random_state = 1)
knn = KNeighborsClassifier(n_neighbors = 3)
x,y = df_parsed.loc[:,df_parsed.columns != 'game_status'], df_parsed.loc[:,'game_status']
knn.fit(x_train,y_train)
prediction = knn.predict(x_test)
#print('Prediction: {}'.format(prediction))
print('With KNN (K=3) accuracy is: ',knn.score(x_test,y_test)) # accuracy
```

```python
# Model complexity
neig = np.arange(1, 25)
train_accuracy = []
test_accuracy = []
# Loop over different values of k
for i, k in enumerate(neig):
    # k from 1 to 25(exclude)
    knn = KNeighborsClassifier(n_neighbors=k)
    # Fit with knn
    knn.fit(x_train,y_train)
    #train accuracy
    train_accuracy.append(knn.score(x_train, y_train))
    # test accuracy
    test_accuracy.append(knn.score(x_test, y_test))

# Plot
plt.figure(figsize=[13,8])
plt.plot(neig, test_accuracy, label = 'Testing Accuracy')
plt.plot(neig, train_accuracy, label = 'Training Accuracy')
plt.legend()
plt.title('-value VS Accuracy')
plt.xlabel('Number of Neighbors')
plt.ylabel('Accuracy')
plt.xticks(neig)
plt.savefig('graph.png')
plt.show()
print("Best accuracy is {} with K = {}".format(np.max(test_accuracy),1+test_accuracy.index(np.max(test_accuracy))))
```
