### Cleaning the data

```python


list_headers = ["Level","Champion","Score","Game Mode","KDA","Creep Score","Gold","Minutes","Date"]

list_stats_read = [line for line in open('LoLMatchHistory.txt', 'r').read().splitlines() if line]
list_stats_cleaned = [line for line in list_stats_read if line != ' ' and line != None]
list_stats_np = np.array_split(np.array(list_stats_cleaned),len(list_stats_cleaned) / 9)
```

```python
df = pd.DataFrame(list_stats_np, columns=list_headers)
df
```

```python
def clean_tab(x):
    x = x.replace("\t", "")
    return x 

df['Level'] = df['Level'].apply(clean_tab) #remove tabs from level_column
df
```

```python
def clean_score(score):
    if score == "Summoner's Rift":
        score = "N/A"
    return score

def sort_scores(list_rank, list_exp, score):
    if(score != "N/A"):
        score_tup = score.split("/")
        if(score_tup[0] and "XP" not in score_tup[0] ):
            list_rank.append(score_tup[0])
        else: 
            list_rank.append("N/A")
            score_tup.append(score_tup[0])
            
        try:
            score_tup_1 = score_tup[1]
        except IndexError:
            score_tup_1 = None
            list_exp.append(0)
        
        if(score_tup_1):
            score_tup_1 = score_tup_1.replace(" XP", "")
            score_tup_1 = score_tup_1.replace("+", "")
            list_exp.append(int(score_tup_1))
    else:
        list_rank.append("N/A")
        list_exp.append(0)
        
    return list_rank, list_exp
    

df['Score'] = df['Score'].apply(clean_score)

list_score = list(df['Score'])

list_rank = []
list_exp = []

count = 0
for score in list_score:
    list_rank, list_exp = sort_scores(list_rank, list_exp, score)
    
df['Rank'] = list_rank
df['Experience Gained'] = list_exp

df
```

```python


list_KDA = list(df['KDA'])

list_split_KDA = [x.split("/") for x in list_KDA]

df['Kills'] = [item[0] for item in list_split_KDA]
df['Deaths'] = [item[1] for item in list_split_KDA]
df['Assists'] = [item[2] for item in list_split_KDA]

df
    
```

```python
def parse_gold(x):
    if('k' in x):
        x = x.replace('k','')
        return float(x) * 1000
    else:
        return float(x)

df.columns = df.columns.str.lower().str.replace(' ', '_')
df['gold'] = df['gold'].apply(parse_gold)

df
```

```python
df.to_csv("CleanLoLMatchHistory3.csv", index=False)
df
```
